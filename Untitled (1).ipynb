```python

# ============================================================
# MARKETING FUNNEL & CONVERSION PERFORMANCE ANALYSIS
# Complete Jupyter Notebook Code
# ============================================================

# STEP 1: IMPORT LIBRARIES
# -------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

sns.set_style("whitegrid")

print("Libraries Imported Successfully")


# ============================================================
# STEP 2: LOAD DATASET
# ============================================================

df = pd.read_csv("marketing_funnel_dataset.csv")

print("Dataset Loaded Successfully")
print(df.head())


# ============================================================
# STEP 3: DATA UNDERSTANDING
# ============================================================

print("\nDataset Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())


# ============================================================
# STEP 4: DATA CLEANING
# ============================================================

df.drop_duplicates(inplace=True)

print("\nData Cleaning Completed")


# ============================================================
# STEP 5: KPI ANALYSIS
# ============================================================

total_visits = df['WebsiteVisits'].sum()
total_leads = df['LeadsGenerated'].sum()
total_customers = df['CustomersAcquired'].sum()
total_revenue = df['Revenue'].sum()
total_spend = df['AdSpend'].sum()

conversion_rate = (total_customers / total_visits) * 100
lead_conversion_rate = (total_customers / total_leads) * 100
roi = ((total_revenue - total_spend) / total_spend) * 100

print("\n========== MARKETING KPIs ==========")

print(f"Total Website Visits: {total_visits}")
print(f"Total Leads Generated: {total_leads}")
print(f"Total Customers Acquired: {total_customers}")
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Ad Spend: ${total_spend:,.2f}")
print(f"Overall Conversion Rate: {conversion_rate:.2f}%")
print(f"Lead-to-Customer Conversion Rate: {lead_conversion_rate:.2f}%")
print(f"ROI: {roi:.2f}%")


# ============================================================
# STEP 6: FUNNEL ANALYSIS
# ============================================================

funnel = [
    total_visits,
    total_leads,
    df['QualifiedLeads'].sum(),
    df['ProposalsSent'].sum(),
    total_customers
]

stages = [
    "Website Visits",
    "Leads",
    "Qualified Leads",
    "Proposals Sent",
    "Customers"
]

plt.figure(figsize=(10,6))

sns.barplot(
    x=stages,
    y=funnel
)

plt.title("Marketing Funnel Analysis")
plt.ylabel("Count")

plt.show()


# ============================================================
# STEP 7: CHANNEL PERFORMANCE ANALYSIS
# ============================================================

channel_revenue = df.groupby('MarketingChannel')['Revenue'].sum()

plt.figure(figsize=(12,6))

sns.barplot(
    x=channel_revenue.index,
    y=channel_revenue.values
)

plt.title("Revenue by Marketing Channel")
plt.xlabel("Marketing Channel")
plt.ylabel("Revenue")

plt.xticks(rotation=15)

plt.show()


# ============================================================
# STEP 8: CONVERSION RATE BY CHANNEL
# ============================================================

channel_conversion = (
    df.groupby('MarketingChannel')['CustomersAcquired'].sum() /
    df.groupby('MarketingChannel')['WebsiteVisits'].sum()
) * 100

plt.figure(figsize=(12,6))

sns.barplot(
    x=channel_conversion.index,
    y=channel_conversion.values
)

plt.title("Conversion Rate by Channel")
plt.ylabel("Conversion Rate (%)")

plt.xticks(rotation=15)

plt.show()


# ============================================================
# STEP 9: REGION PERFORMANCE ANALYSIS
# ============================================================

region_customers = df.groupby('Region')['CustomersAcquired'].sum()

plt.figure(figsize=(10,6))

sns.barplot(
    x=region_customers.index,
    y=region_customers.values
)

plt.title("Customers Acquired by Region")
plt.xlabel("Region")
plt.ylabel("Customers")

plt.show()


# ============================================================
# STEP 10: DEVICE ANALYSIS
# ============================================================

device_conversion = (
    df.groupby('DeviceType')['CustomersAcquired'].sum() /
    df.groupby('DeviceType')['WebsiteVisits'].sum()
) * 100

plt.figure(figsize=(8,5))

sns.barplot(
    x=device_conversion.index,
    y=device_conversion.values
)

plt.title("Device-wise Conversion Rate")
plt.ylabel("Conversion Rate (%)")

plt.show()


# ============================================================
# STEP 11: REVENUE VS AD SPEND
# ============================================================

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x='AdSpend',
    y='Revenue',
    hue='MarketingChannel'
)

plt.title("Revenue vs Ad Spend")

plt.show()


# ============================================================
# STEP 12: MONTHLY TREND ANALYSIS
# ============================================================

df['CampaignDate'] = pd.to_datetime(df['CampaignDate'])

df['Month'] = df['CampaignDate'].dt.to_period('M')

monthly_revenue = df.groupby('Month')['Revenue'].sum()

plt.figure(figsize=(14,6))

monthly_revenue.plot(marker='o')

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()


# ============================================================
# STEP 13: DROP-OFF ANALYSIS
# ============================================================

drop_offs = []

for i in range(len(funnel)-1):
    drop = ((funnel[i] - funnel[i+1]) / funnel[i]) * 100
    drop_offs.append(drop)

for i in range(len(drop_offs)):
    print(f"Drop-off from {stages[i]} to {stages[i+1]}: {drop_offs[i]:.2f}%")


# ============================================================
# STEP 14: CORRELATION HEATMAP
# ============================================================

numeric_df = df[['WebsiteVisits','LeadsGenerated',
                 'QualifiedLeads','ProposalsSent',
                 'CustomersAcquired','AdSpend','Revenue']]

plt.figure(figsize=(10,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()


# ============================================================
# STEP 15: TOP CHANNELS
# ============================================================

top_channel = channel_revenue.idxmax()

print("\n========== KEY INSIGHTS ==========\n")

print(f"1. Highest revenue generated by: {top_channel}")

best_conversion_channel = channel_conversion.idxmax()

print(f"2. Best conversion channel: {best_conversion_channel}")

highest_region = region_customers.idxmax()

print(f"3. Highest customer acquisition region: {highest_region}")

highest_device = device_conversion.idxmax()

print(f"4. Highest converting device type: {highest_device}")


# ============================================================
# STEP 16: BUSINESS RECOMMENDATIONS
# ============================================================

print("\n========== RECOMMENDATIONS ==========\n")

print("1. Increase budget allocation to high-converting channels.")
print("2. Improve landing pages to reduce funnel drop-offs.")
print("3. Retarget abandoned leads through email campaigns.")
print("4. Optimize mobile user experience for better conversions.")
print("5. Focus on high-performing regions for customer acquisition.")


# ============================================================
# STEP 17: EXPORT CLEANED DATA
# ============================================================

df.to_csv("cleaned_marketing_funnel_data.csv", index=False)

print("\nCleaned dataset exported successfully.")


# ============================================================
# STEP 18: PROJECT COMPLETED
# ============================================================

print("\n========================================")
print(" MARKETING FUNNEL ANALYSIS COMPLETED ")
print("========================================")
```
